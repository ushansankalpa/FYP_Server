```python
import numpy as np
import pandas as pd
import sklearn.metrics as sm
```

```python
dataset = pd.read_csv('SL_csv.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop(columns =['Gender','Age'])
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Learner'] = le.fit_transform(dataset['Learner'])
```

```python
x = dataset.drop('Learner', axis=1)
y = dataset['Learner']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.25)
```

```python
# X_train = X_train.shape(X_train.shape[0])
X_train.shape , X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, Y_train)
    acc = model.score(X_test,Y_test)
    print(str(model)+ ' ----> '+ str(acc))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
model_acc(rfc)
# rfc.fit(X_train, Y_train)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10,50,100],
             'criterion':['squared_error','absolute_error', 'poisson']}

grid_obj = GridSearchCV(estimator = rf , param_grid = parameters)
grid_fit = grid_obj.fit(X_train, Y_train)

best_model = grid_fit.best_estimator_
best_model

# new_data = [[1, 1, 1, 1, 1, 1, 1, 1, 1, 3, 5, 5, 1, 5, 5]]
# prediction = rf.predict(new_data)
# print(prediction)
```

```python
best_model.score(X_train,Y_train)
```
